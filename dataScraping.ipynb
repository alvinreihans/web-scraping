#### Mengimport library

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

#### Mengambil halaman MyAnimeList Top Anime

```python
url = 'https://myanimelist.net/topanime.php'


try:
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  anime_list = soup.findAll('tr', class_='ranking-list')
  print(soup)

except Exception as e:
  print(e)
```

#### Mengambil data judul anime

```python
title_columns = []

for e in anime_list:
  title_element = e.find('h3', class_='fl-l fs14 fw-b anime_ranking_h3')
  if title_element:
    title_columns.append(title_element.text.strip())

print(title_columns)
```

#### Mengambil data rating anime

```python
rating_elements = []
for e in anime_list:
  rating_elements.append(str(e.find('div', class_='js-top-ranking-score-col di-ib al')))

html_content = ''.join(rating_elements)
soup = BeautifulSoup(html_content, 'html.parser')
score_divs = soup.find_all('div', class_='js-top-ranking-score-col')
rating_columns = [float(div.find('span', class_='score-label').text) for div in score_divs]
print(rating_columns)
```

#### Mengambil data informasi anime

```python
html_content = str(anime_list)
soup = BeautifulSoup(html_content, 'html.parser')
anime_info_elements = soup.find_all(class_="information di-ib mt4")
anime_info_columns = [element.get_text(strip=True) for element in anime_info_elements]
print(anime_info_columns)
```

#### Menampilkan hasil data scraping menggunakan pandas

```python
anime_count = len(anime_list)
table = {'Rank': np.arange(1, anime_count + 1),
         'Title': title_columns,
         'Rating': rating_columns,
         'Info': anime_info_columns}
df = pd.DataFrame(table)
# df.reset_index(drop=True, inplace=True)
# df.index = df.index + 1
# df.index.name = 'Rank'
df.style.hide()
```
